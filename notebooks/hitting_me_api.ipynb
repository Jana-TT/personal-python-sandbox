```python
import httpx

params = {'latitude': 29.742180, 'longitude': -95.487450}
r = httpx.get("https://api.open-meteo.com/v1/forecast?&hourly=temperature_2m&temperature_unit=fahrenheit", params=params)
r.text
```

```python
r_json = r.json()
r_json
```

```python
import polars as pl

testing = pl.LazyFrame(r_json["hourly"])
testing.collect()
```

```python
poop = testing.select("time").collect()
poop
```

```python
shit_2 = testing.select(pl.col("time").cast(pl.Datetime))
shit_2.collect()
```

```python
#he didnt like this way
shit3 = testing.with_columns(pl.col("time").str.strptime(pl.Datetime, strict=False))
shit3.collect()
```

```python
poop4 = testing.select(pl.col("time").str.to_datetime("%Y-%m-%dT%H:%M"), pl.col("temperature_2m"))
poop4.collect()
```
